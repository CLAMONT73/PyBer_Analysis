```python
# Import dependencies.

import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.

x_axis = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floatsas the total fare $ accumulated for each month.

y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot using the MATLAB method.

plt.plot(x_axis, y_axis)
```

```python
# Create the plot using Object Oriented Interface method.

fig, ax = plt.subplots()

ax.plot(x_axis, y_axis)
```

```python
# The above can also be written like this. The first two lines return an empty chart.
# Adding ax.plot() returns the same chart as above.

fig = plt.figure()

ax = fig.add_subplot()

ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt().

ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot using plt.show() / Use once per session, otherwise it can cause the graph to display
# improperly. If issues with the function, restart the kernel & clear the output: Restart & Clear Output.

plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot and add a label.

plt.plot(x_axis, y_axis, label='Boston')

# Create x & y axes labels.

plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45.

plt.ylim(0, 45)

# Add a title.

plt.title("PyBer Fare by Month")

# Add a legend.

plt.legend()
```

```python
# Adding line formatting, including marker.

plt.figure(figsize=(20, 10))

plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')

plt.xlabel("Date")
plt.ylabel("Fare($)")

plt.xlim(0, 5)
plt.ylim(0, 45)

plt.title("PyBer Fare by Month")

plt.grid()

plt.legend()
```

```python
# Some markers are too small. Make them bigger by multiplying them by any value. To do this, iterate through
# the y_axis data and multiply each point by 3 and add it to a new list.

y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
    
plt.scatter(x_axis, y_axis, s=y_axis_larger)    
```

```python
# Using the LIST COMPREHENSION technique to replace for and while loops.
# Standard Format: new_list = [expression for item in list if conditional]

plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
'''SKILL DRILL: Using the Object-Oriented method, make the following changes to the bubble chart:
Make markers sky blue, change marker size to 5x each data point, make color 20% transparent, add black edge
to circles, make linewidth of circles 2 points, add legend for the city of Boston, add title, switch axis so
Fare($) data is on the x-axis, add limit for the x-axis (0-50), & invert y-axis so January is on top'''

fig, ax = plt.subplots()

ax.scatter(y_axis, x_axis, c="skyblue", label="Boston", edgecolors="black", linewidth=2, s = [i * 5 for i in y_axis])

ax.invert_yaxis()

ax.MarkerFaceAlpha = 0.2
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")

ax.set_xlim(0, 50)

ax.legend()
```

```python
# Create a pie chart using plt.pit() & MATLAB method. Use x-axis data (months) as labels & y-axis data as values.

plt.pie(y_axis, labels=x_axis)

plt.show()
```

```python
# Apply the "explode" or "pop out" parameter to show the largest %. "0" is the distance from the center of the
# pie, and "1" is completely outside of the pie. Assign "1" to the largest % which is in the month of July.

plt.subplots(figsize=(8, 8))

explode_values =(0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)

plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')

plt.show()
```

```python
# Assigning different colors.

colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]

explode_values =(0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))

plt.pie(y_axis, 
        explode=explode_values, 
        colors=colors, 
        labels=x_axis, 
        autopct='%.1f%%')

plt.show()
```
