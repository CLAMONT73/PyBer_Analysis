```python
# Import dependencies. The "statistics" module comes with the Anaconda istallation. Need it to add y-error bars
# to line and bar charts, and calculate STDEV of the fare amount.

import matplotlib.pyplot as plt
import statistics
```

```python
# Set the x-axis to a list of strings for each month.

x_axis = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floatsas the total fare $ accumulated for each month.

y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
stdev = statistics.stdev(y_axis)

stdev
```

```python
# Add a cap to error bars with the capsize= parameter.

plt.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
# Add error bars and a capsize using the Object-Oriented method.

fig, ax = plt.subplots()

ax.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)

plt.show()
```

```python
# Adding error bars to a bar chart using the MATLAB method.

plt.bar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
# Adding error bars to a bar chart using the Object-Oriented method.

fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, yerr=stdev, capsize=3)
plt.show()
```

```python
# Create a horizontal bar chart.

plt.barh(x_axis, y_axis)
```

```python
# Use the xticks() function to adjust the major x-axis ticks on the horizontal bar. This is the spacing of the
# values in the x-axis. Default: Every $10 in the above chart. Change it to $5.

import numpy as np

plt.barh(x_axis, y_axis)

plt.xticks(np.arange(0, 51, step=5.0))

plt.gca().invert_yaxis()
```

```python
# Numpy's np.arange() function: If "step" is not included, function returns all #'s in the specified range.

np.arange(0, 51, step=5.0)
```

```python
# Creat the same graph using the Object-Oriented method.

fig, ax = plt.subplots()

ax.barh(x_axis, y_axis)

ax.set_xticks(np.arange(0, 51, step=5.0))

plt.show()
```

```python
# Use the set_minor_locator function with the xaxis class in the Object-Oriented method to add minor ticks without
# numberical values between the x & y axes. Inside the parentheses, use the MultipleLocator function.
# Standard Format: x.axis.set_minor_locator().

from matplotlib.ticker import MultipleLocator

# Increase plot size.

fig, ax = plt.subplots(figsize=(8, 8))

ax.barh(x_axis, y_axis)

ax.set_xticks(np.arange(0, 51, step=5.0))

# Create minor ticks at increments of 1.

ax.xaxis.set_minor_locator(MultipleLocator(1))

plt.show()
```
